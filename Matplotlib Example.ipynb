```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
% matplotlib notebook
```

```python
df = pd.DataFrame(np.random.randn(50,4), columns=["a","b","c","d"])
df.head(10)
```

```python
df.plot.scatter(x='a',y='b', c='r', s=(df['b']*100, df['a']*100), label= 'Interaction Between A and B', title= 'My Title \n' )
```

```python
df.plot.bar()
plt.show()
```
